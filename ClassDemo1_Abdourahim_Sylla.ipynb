```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing data from a website link using csv
videos = pd.read_csv("https://raw.githubusercontent.com/fedhere/FDSFE_FBianco/main/data/USyoutubes.csv")
```

```python
#Showing the videos from the start/first one
videos.head()
```

```python
#Restricting the videos to only display the views, likes, dislikes and comments
videos_restrict = videos[["views","likes","dislikes","comment_count"]]
```

```python
#Seeing the result from the previous code
videos_restrict.head()
```

```python
#Same result, but from the end
videos_restrict.tail()
```

```python
#Displaying the highest number of likes
videos["likes"].max()
```

```python
#Displaying the minimum amount of likes
videos["likes"].min()
```

```python
#Most likes with video information details
max_likes = videos_restrict["likes"].max()

videos[videos["likes"] == max_likes]
```

```python
#Most dislikes with video information details
max_dislikes = videos_restrict["dislikes"].max()

videos[videos["dislikes"] == max_dislikes]
```

```python
#Least likes with video information details
least_likes = videos_restrict["likes"].min()

videos[videos["likes"] == least_likes]
```
